```python
# monet 4
# van goch 22
# landscape 4

with open("/datasets/ee285f-public/wikiart/Artist/artist_train") as f:
    arts = f.readlines()[2:]
with open("/datasets/ee285f-public/wikiart/Artist/artist_val") as f:
    arts += f.readlines()[2:]
    
print(len(arts))
```

```python
with open("/datasets/ee285f-public/wikiart/Genre/genre_train.csv") as f:
    genre = f.readlines()
with open("/datasets/ee285f-public/wikiart/Genre/genre_val.csv") as f:
    genre += f.readlines()
    
print(len(genre))
```

```python
landscape = []
for gen in genre:
    split_gen = gen.split(',')
    
    if split_gen[1] == '4\n':
        landscape.append(split_gen[0])
```

```python
vg = []
for art in arts:
    split_art = art.split(",,")
    
    if split_art[1] == '22\n':
        vg.append(split_art[0])
    
```
